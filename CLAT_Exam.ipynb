```python
import gradio as gr

# Rule-based CLAT chatbot function
def clat_chatbot(user_input):
    user_input = user_input.lower()

    if "exam date" in user_input or "when is clat" in user_input:
        return "üìÖ CLAT is usually conducted in December every year. Check the official site for exact dates."

    elif "eligibility" in user_input:
        #return "‚úÖ To be eligible for CLAT UG, you must have passed 10+2 with at least 45% marks (40% for SC/ST)."
        return "There is no upper age limit for candidates appearing for UG Programme through the CLAT 2025. \
         Candidates who have passed in 10+2 or an equivalent examination with a minumum of:\n \
         Forty five percent (45%) of marks or its equivalent in grade or \n \
         Forty Percent (40%) of marks or equivalent in case of candidates belonging to SC/ST/PwD categories.\n \
         Candidates appearing for the qualifying examination in March/ April, 2025 are also eligible to apply.\
         However, they shall be required to produce an evidence of their passing the qualifying examination at the time of admission,\
        failing which they shall lose their right to be considered for admission.\
          The result of the qualifying examination (i.e., 10+2) shall be submitted by the candidate at the time of admission, \
          failing which the candidate shall be ineligible for admission to the course they may have applied for."

    elif "syllabus" in user_input:
      return "üìòThe syllabus includes English, Current Affairs, Legal Reasoning, Logical Reasoning, and Quantitative Techniques."

    elif "application" in user_input or "how to apply" in user_input:
        return "üìù You can apply through consortiumofnlus.ac.in. Applications typically open in August."

    elif "exam pattern" in user_input:
        return "üß† CLAT UG has 150 MCQs, 2 hours duration. +1 for correct, -0.25 for wrong answers."

    elif "college" in user_input or "nlu" in user_input:
        return "üèõÔ∏è There are 22 NLUs including NLSIU Bangalore, NALSAR Hyderabad, NUJS Kolkata, and more."

    elif "english questions" in user_input or "how many english questions" in user_input:
        return " There are 22 to 26 English questions in CLAT exam."

    elif "preparation" in user_input or "how to prepare" in user_input:
        return "üìñ Read newspapers, take mock tests, study basics, and practice time management. Join a test series if possible."

    elif "last year nlsiu bangalore" in user_input:
        return "üìñ In 2023 NLSIU Bangalore was 114 for the General category, according to Toprankers."

    elif "cutoff" in user_input or "rank" in user_input:
        return "üéØ CLAT cutoffs vary. Top NLUs need ranks within top 100-150 for General category."

    else:
        return "ü§î I'm not sure about that. Ask me about CLAT syllabus, eligibility, colleges, or preparation tips!"

# Launch Gradio interface
iface = gr.Interface(fn=clat_chatbot,
                     inputs="text",
                     outputs="text",
                     title="CLAT Exam Chatbot",
                     description="Ask questions about the Common Law Admission Test (CLAT)",
                     theme="default")

iface.launch()
```
